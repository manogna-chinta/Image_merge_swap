```python
import matplotlib.pyplot as plt
from PIL import Image
import time

def render_img(img):
    image = Image.open(img)
    plt.imshow(image)
    plt.show()
    return image

def auto_resize(img_1, img_2):
    master_size = img_1.size
    img_2.resize((master_size[0], master_size[1]))
    return master_size

def img_merge(img_1, img_2, master_size):
    merged_image = Image.new("RGB", (master_size[0] * 2, master_size[1]), (250, 250, 250))
    merged_image.paste(img_1)
    merged_image.paste(img_2, (master_size[0], 0))
    output_path = "merged.png"
    merged_image.save(output_path)
    render_img(output_path)

if __name__ == "__main__":
    print("Running image_merger.py as main script...")
    time.sleep(1)
    
    print("Displaying image1.png...")
    img_1 = render_img("image1.png")
    time.sleep(2)
    
    print("Displaying image2.png...")
    img_2 = render_img("image2.png")
    time.sleep(2)
    
    print("Merging image1.png and image2.png...")
    print("Merged image saved as merged.png")
    time.sleep(1)
    
    (img_merge(img_1, img_2, auto_resize(img_1, img_2)))
    time.sleep(1)
    
    print("Exiting image_merger.py...")
```
